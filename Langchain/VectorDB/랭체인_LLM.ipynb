### __LLM__
* __LLM (Text Generation)__: 텍스트를 입력받아 다음 단어를 예측하고 문맥을 파악해 글을 생성합니다. 디코더(Decoder) 중심의 구조를 가집니다 (Large Language Model (LLM) is a massive, generative model designed to understand context and generate natural language text, code, or image)
* __Embedding Model (Vector Conversion)__: 텍스트를 고차원 숫자 벡터로 변환합니다. 변환된 벡터들은 의미가 비슷할수록 벡터 공간상에 가깝게 배치되며, 주로 인코더(Encoder) 중심의 구조를 띱니다 (An Embedding Model is designed to translate data (like text) into numerical lists (vectors) that represent semantic meaning, making it ideal for similarity searches)

### __RAG (Retrieval-Augmented Generation)__
* __an AI framework that improves Large Language Models (LLMs) by pulling facts from an external knowledge base before generating a respons__
* __Data Ingestion & Indexing__ : 외부 지식 소스를 효과적으로 검색할 수 있도록 미리 준비하는 단계대규모 문서 집합(예: 위키백과, 뉴스 데이터, 논문 등)을 수집각 문서를 작은 청크(예: 문장, 단락)로 나누고, 개별 벡터 임베딩으로 변환벡터 DB(예: FAISS, ANNoy, Elasticsearch 등)에 저장
    * __Embedding__:  chunking is the process of breaking down large text into smaller segments called chunks. These text chunks are converted into numerical vectors (embeddings) that capture their semantic meaning
    * __Storage__ : The vectors are saved in a vector database (such as Chroma, Pinecone, or Milvus) so the system can search them quickly
* __User Input__ : 사용자의 질의(질문, 문장 등)를 모델이 처리할 수 있는 형태로 변환사용자가 자연어 형태로 질문을 입력 (예: “Who discovered penicillin?”)입력된 텍스트를 벡터 임베딩(예: BERT, DPR, OpenAI embedding 등)으로 변환 (When a user submits a prompt, the system converts their query into a vector)
* __Retrieval__ : 사용자 입력과 가장 관련성이 높은 문서를 색인에서 찾아내어, 변환된 사용자 입력 벡터와 색인된 문서 벡터 간 유사도 계산 (ex: 코사인 유사도, dot product 등)가장 유사한 문서 상위 N개(예: 5~10개) 검색 -> 반환검색된 문서를 RAG 모델의 컨텍스트로 사용 (Search and fetch relevant informatiojn from the knowledge base)
* __Generation__ : 검색된 문서를 기반으로 최종 답변을 생성 (The LLM uses the attached external context to answer the user's question, significantly reducing the chance of hallucinations while keeping responses highly accurate) - Generate the response using an LLM based on the augmented user question and retrieved information
